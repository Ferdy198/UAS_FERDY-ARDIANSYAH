```python
class WhileLoopCompiler:
    def __init__(self, source_code):
        self.source_code = source_code
        self.label_counter = 1

    def new_label(self):
        lbl = f"L{self.label_counter}"
        self.label_counter += 1
        return lbl

    def lexical_analysis(self):
        tokens = self.source_code.replace('(', ' ( ').replace(')', ' ) ').replace('{', ' { ').replace('}', ' } ').split()
        return tokens

    def syntax_semantic_analysis(self, tokens):
        if 'while' in tokens and '{' in tokens and '}' in tokens:

            condition = " ".join(tokens[tokens.index('(')+1 : tokens.index(')')])
            

            body = " ".join(tokens[tokens.index('{')+1 : tokens.index('}')])
            
            return condition, body
        else:
            raise SyntaxError("Struktur Perulangan While tidak valid.")

    def generate_tac(self):
        tokens = self.lexical_analysis()
        condition, body = self.syntax_semantic_analysis(tokens)

        label_start = self.new_label()
        label_else = self.new_label() 

        tac = []
        tac.append(f"{label_start}:")
        tac.append(f"ifFalse {condition} goto {label_else}")
        tac.append(f"    {body}")
        tac.append(f"    goto {label_start}")
        tac.append(f"{label_else}:")
        
        return "\n".join(tac)
    

#PENGGUANAAN NYA

source = "while ( x < 10 ) { x = x + 1 }"
compiler = WhileLoopCompiler(source)

print("=== PROGRAM SIMULASI TEKNIK KOMPILASI (VERSI PERULANGAN WHILE) ===")
print(f"Source Code Input: {source}\n")

#Tahap Leksikal
tokens_result = compiler.lexical_analysis()
print("ANALISIS LEKSIKAL (Tokens)")
print(tokens_result)
print()

#Tahap Sintaksis & Semantik
cond, body_loop = compiler.syntax_semantic_analysis(tokens_result)
print("TAHAP ANALISIS SINTAKSIS & SEMANTIK")
print(f"Kondisi Loop     : [ {cond} ]")
print(f"Isi/Body Loop    : [ {body_loop} ]")
print("Status           : VALID")
print()

#Hasil Akhir TAC
tac_result = compiler.generate_tac()
print("GENERASI THREE-ADDRESS CODE (TAC)")
print(tac_result)
```
